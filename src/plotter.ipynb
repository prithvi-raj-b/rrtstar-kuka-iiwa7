```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

## Plotter for ED5215

```python
with open("../data1.csv",newline='') as file:
    reader = csv.reader(file, delimiter=',')
    next(reader)
    data1 = list(reader)

rt, rn, rp, st, sn, sp, bt, bn, bp = [], [], [], [], [], [], [], [], []
for row in data1:
    if row[0] == 'r':
        rt.append(float(row[1]))
        rn.append(int(row[2]))
        rp.append(float(row[3]))
    elif row[0] == 's':
        st.append(float(row[1]))
        sn.append(int(row[2]))
        sp.append(float(row[3]))
    elif row[0] == 'b':
        bt.append(float(row[1]))
        bn.append(int(row[2]))
        bp.append(float(row[3]))

rr = [1000*x/y for x,y in zip(rt,rn)]
sr = [1000*x/y for x,y in zip(st,sn)]
br = [1000*x/y for x,y in zip(bt,bn)]

categories = ['RRT', 'RRT*', 'Bi-Directional RRT']

tavg = [np.mean(rt), np.mean(st), np.mean(bt)]
navg = [np.mean(rn), np.mean(sn), np.mean(bn)]
pavg = [np.mean(rp), np.mean(sp), np.mean(bp)]
ravg = [np.mean(rr), np.mean(sr), np.mean(br)]

terr = [max(rt)-min(rt), max(st)-min(st), max(bt)-min(bt)]
nerr = [max(rn)-min(rn), max(sn)-min(sn), max(bn)-min(bn)]
perr = [max(rp)-min(rp), max(sp)-min(sp), max(bp)-min(bp)]
rerr = [max(rr)-min(rr), max(sr)-min(sr), max(br)-min(br)]
```

```python
plt.bar(categories, tavg, yerr=terr, capsize=5)
plt.xlabel('Algorithm')
plt.ylabel('Time (s)')
```

```python
plt.bar(categories, navg, yerr=nerr, capsize=5)
plt.xlabel('Algorithm')
plt.ylabel('No. of Nodes')
```

```python
plt.bar(categories, pavg, yerr=perr, capsize=5)
plt.xlabel('Algorithm')
plt.ylabel('Path Length')
```

```python
plt.bar(categories, ravg, yerr=rerr, capsize=5)
plt.xlabel('Algorithm')
plt.ylabel('Node Creation Time (ms)')
```
